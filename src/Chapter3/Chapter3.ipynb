## 3.2節：基本変形

```julia
using LinearAlgebra
using SparseArrays
```

```julia
# 基本変形 P
function P(m,i,j)
    A = spdiagm(0 => ones(m)) # diagmでもOK (spdiagmを用いるにはSparseArraysが必要)
    A[i, :], A[j, :] = A[j, :], A[i, :]
    return A
end

P(6,2,5)
```

```julia
P(6,2,5)*P(6,2,5)
```

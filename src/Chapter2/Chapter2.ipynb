```julia
using Plots
using LinearAlgebra # diagmを用いるために必要
using LaTeXStrings

x_ref = 0:1/1000:1 # range(0, 1, length=1000) のように書いてもよい
g_ref = 2*pi*cos.(2*pi*x_ref) # 解析的な導関数の値（f'(x) = 2πcos(2πx)）を計算
plot(x_ref,g_ref,color=:black,xlabel = "x",label=L"f\prime (x) = 2\pi \cos (2\pi x) ") # 基準の導関数グラフを黒色でプロット，x軸ラベルを"x"，凡例にLaTeXで数式を表示

function derivative_approx(n)
    Δx = 1/n
    x = 0:Δx:1
    f = sin.(2*pi*x) # 関数f(x) = sin(2πx)の値を計算
    A = 1/(2*Δx) * diagm(1 => ones(n-2),-1 => -ones(n-2)) # 中心差分法による1階導関数近似のための差分行列を構成
    g = A * f[2:n] # 内部ノードに対して差分行列を適用して導関数を近似
    return x[2:n], g # x座標と近似導関数の値を返す（端点を除く）
end

# Δx = 0.1のときの近似導関数を白色の四角マーカーでプロット（既存のグラフに追加）
plot!(derivative_approx(10),seriestype=:scatter,markershape=:rect,markercolor=:white,
    label=L"\Delta x = 0.1")
# Δx = 0.05のときの近似導関数を灰色の円マーカーでプロット（既存のグラフに追加）
plot!(derivative_approx(20),seriestype=:scatter,markershape=:circ,markercolor=:gray,
    label=L"\Delta x = 0.05")
```

### 1.4.3節：制御構文

```julia
(x,y) = (3,2)
if x < y
    println("x is less than y")
elseif x > y
    println("x is greater than y")
else
    println("x is equal to y")
end
```

```julia
for i = 1:4
    print("i = $i, ") # iの値を出力
end
```

```julia
for i = [-4, 5, 2]
    print("i = $i, ") # iの値を出力
end
```

```julia
i = 1
while i < 5
    print("i = $i, ") # iの値を出力
    i += 1 # iの値に1を加えたものを新たにiとする
end
```

```julia
for i = 1:5
    print("i = $i, ") # iの値を出力
    if i == 3
        break
    end
end
```

```julia
for i = 1:5
    if i % 2 != 0
        continue
    end
    print("i = $i, ")
end
```

### 1.4.4節：関数

```julia
function mysum(n)
    s = 0
    for i = 1:n
        s += i
    end
    return s
end
```

```julia
sum(10)
```

```julia
sum(1:10)
```

```julia
sum(3:10)
```

```julia
function mysum_range(i, j)
    s = 0
    for k = i:j
        s += k
    end
    return s
end
```

```julia
mysum_range(3,10)
```

```julia
function f(x;y=1)
    return x + y
end
```

```julia
f(2)
```

```julia
f(2, y=0.1)
```

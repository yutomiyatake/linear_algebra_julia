## 4.2節：行列式の定義

```julia
# 置換の例
σ = Permutation([3, 2, 1])
```

```julia
# 置換の例
σ = Permutation([2, 4, 3, 1])
```

```julia
# 置換の積
σ = Permutation([2, 3, 1])
τ = Permutation([3, 2, 1])
σ * τ
```

```julia
τ * σ   
```

```julia
# 逆置換の例
σ = Permutation([2, 3, 1])
inv(σ)
```

```julia
# 置換の符号
σ = Permutation([2, 4, 1, 3])
sign(σ)
```

## 4.3節：行列式の諸性質・公式

```julia
A = Rational.([2 3 -1 10; -1 -6 -3 -7; -3 -9 -10 -8; 1 3 2 5]) 
    # Rational. をつけないと，整数型になり，続く計算では分数が現れたときに小数表示に
    # なってしまうため，見やすさのために，有理数型にしておく．
    # A = [2//1 3 -1 10; -1 -6 -3 -7; -3 -9 -10 -8; 1 3 2 5] のように書いても同じ
```

```julia
L, U, p = lu(A)
```

```julia
p # 置換を表すベクトル
```

```julia
sign(Permutation(p)) # pの符号（Permutationsパッケージを利用）
```

```julia
sign(Permutation(p)) * det(U)
```

```julia
det(A)
```

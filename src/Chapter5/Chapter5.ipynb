シュミットの正規直交化法を用いたQR分解

```julia
function my_qr(A)
    m, n = size(A)
    Q = similar(A, m, n)
    R = zeros(eltype(A), n, n)

    for j in 1:n
        # 列ベクトル v を用意
        v = copy(A[:, j])

        # 既に求めた Q の列で射影を引く
        for i in 1:j-1
            R[i, j] = dot(Q[:, i], v)
            v .-= R[i, j] .* Q[:, i]
        end

        # 正規化して Q の列 j を確定
        R[j, j] = norm(v)
        Q[:, j] .= v ./ R[j, j]
    end

    return Q, R
end
```

```julia
A = randn(4,3)
Q, R = my_qr(A)
A - Q * R
Q
```

```julia
A - Q * R
```

```julia
# Aの列が線形従属の場合
A = [1. 2 3; 4 5 9; 7 8 15] # 1列目 + 2列目 = 3列目
Q, R = my_qr(A)
Q
```

```julia
# Aの列が線形従属の場合
A = [1 2 3; 4 5 9; 7 8 15] # 1列目 + 2列目 = 3列目
Q, R = qr(A)
Matrix(Q)
```

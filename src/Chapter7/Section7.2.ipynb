# 第7.2節：力学系（常微分方程式）

バージョン1.10.10で動作確認しています．

事前にインストールが必要なパッケージ
+ LinearAlgebra
+ Plots
+ LaTeXStrings

```julia
using LinearAlgebra, Plots, LaTeXStrings
```

```julia
ω = 100 # 角振動数
x0 = [1.0; 1.0] # 初期値
tspan = (0.0, 1.0) # 時間区間
A = [0.0 1.0; -ω^2 0.0] # 線形項の行列
g(x) = [0; sin(x[1])] # 非線形項
```

```julia
# 陽的オイラー法
N = 100000 # 分割数
Δt = (tspan[2] - tspan[1]) / N # 時間刻み幅
t_explicit = 0:Δt:tspan[2] # 時間点の配列
x_explicit = zeros(2, N+1) # 解の配列
x_explicit[:, 1] = x0 # 初期値の設定
for n in 1:N
    x_explicit[:, n+1] = 
      x_explicit[:, n] + Δt * (A * x_explicit[:, n] + g(x_explicit[:, n]))
end
```

```julia
plot(t_explicit, x_explicit[1, :], 
    xlabel=L"t", ylabel=L"x_1(t)", label=:false, title=L"\Delta t = 10^{-5}", 
    color = :grays, xlabelfontsize=14, ylabelfontsize=14, tickfontsize=10, legendfontsize=12)
# savefig("explicit_euler_1e-5.pdf")
```

```julia
# 指数関数型オイラー法
N = 6 # 分割数
Δt = (tspan[2] - tspan[1]) / N # 時間刻み幅
t_exponential = 0:Δt:tspan[2] # 時間点の配列
x_exponential = zeros(2, N+1) # 解の配列
x_exponential[:, 1] = x0 # 初期値の設定
for n in 1:N
    x_exponential[:, n+1] = 
      exp(Δt * A) * x_exponential[:, n] + A \ (exp(Δt * A) - I) * g(x_exponential[:, n])
end
```

```julia
plot(t_explicit, x_explicit[1, :], 
    label="Explicit Euler", xlabel=L"t", ylabel=L"x_1(t)", 
    color = :grays, xlabelfontsize=14, ylabelfontsize=14, tickfontsize=10, legendfontsize=12)
plot!(t_exponential, x_exponential[1, :], 
    marker=:circle, label="Exponential Euler", color = :grays, linestyle=:dash)
# savefig("exponential_euler_vs_explicit_euler.pdf")
```

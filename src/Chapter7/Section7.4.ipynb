## 7.4.1節：特異値

```julia
Random.seed!(0)
A = randn(4,3)
```

```julia
F = svd(A)
# U, S, V = svd(A) # このように書いてもよい
```

```julia
A ≈ F.U * Diagonal(F.S) * F.V' 
    # U*Σ*V' が A に近いことを確認（≈を使うと true/false で判定可能）
# A ≈ U * Diagonal(S) * V'
```

```julia
# 複素行列の場合
A = rand(4,3) + im * rand(4,3) # 複素数行列
F = svd(A)
# A ≈ F.U * Diagonal(F.S) * F.V'
```

## 7.4.2節：低ランク近似

```julia
using Images, LinearAlgebra, Plots
# using TestImages # サンプル画像を読み込んで使う場合
```

```julia
image = load("osaka_castle.png") # 画像はこのipynbファイルと同じフォルダにおいてください
# image = Gray.(image) # 必要に応じてグレー画像に変換
```

```julia
A = channelview(image) # 2次元配列（行列）に変換
```

```julia
F = svd(A) #特異値分解
# 特異値の分布を降順にプロット
plot(F.S,yaxis=:log,ylabel="singular values", 
    marker=:circle, color=:grays, 
    xlabel="index",legend=false,xlabelfontsize=14,ylabelfontsize=14,tickfontsize=10) 
# savefig("osaka_castle_singular_values.pdf") # 特異値のプロットを保存
```

```julia
n=200 # 大きい方からn個の特異値を使って低ランク近似
U, S, V = F
u = U[:,1:n]
s = S[1:n]
v = V[:,1:n]
An = u * Diagonal(s) * v'
```

```julia
Gray.(An) # 画像として表示
```

```julia
An = clamp.(An,0,1) # 画素値の範囲を[0,1]に収める．低ランク行列には負の値や1より大きなが含まれる可能性があるため．
# save("osaka_castle_200.png",Gray.(An))
```

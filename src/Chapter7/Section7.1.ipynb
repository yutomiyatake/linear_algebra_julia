```julia
# 積み残しの例
function Basels_problem(K)
    x = 0.0
    for k = 1:K
        x += 1/k^2
    end
    return x
end
```

```julia
println(Basels_problem(94906264))
println(Basels_problem(94906265))
println(Basels_problem(94906266))
```

## 行列のノルム

```julia
using LinearAlgebra
```

```julia
A = [1 1; 1 1.001]
```

```julia
println(opnorm(A))
println(opnorm(A,1))
println(opnorm(A,2))
println(opnorm(A,Inf))
```

```julia
A = [1 1; 1 1.001]
```

```julia
println(cond(A))
println(cond(A,1))
println(cond(A,2))
println(cond(A,Inf))
```
